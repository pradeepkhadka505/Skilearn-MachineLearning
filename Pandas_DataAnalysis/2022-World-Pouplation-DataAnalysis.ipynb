### Country having higest populations using sort method

```python
world_pop.sort_values(by='World Population Percentage', ascending=False).head(10)
```

```python
world_pop.sort_values(by='Growth Rate', ascending=False).head(5)
```

#### Continent Grows faster 

```python
world_pop.groupby('Continent').sum().sort_values(by='World Population Percentage', ascending=False)
```

```python
growth_rate = world_pop.groupby('Continent').sum().sort_values(by='World Population Percentage', ascending=False)
```

```python

plt.plot(growth_rate, marker='o')
plt.xticks(rotation= 'vertical')
```

```python
# Creating the line plot
plt.figure(figsize = (15, 6))
sbn.lineplot(data = world_pop.transpose()) # Transpose to change the axis
plt.title("Population Growth by Continent (1970-2022)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(title = "Continent")
plt.show()
```
